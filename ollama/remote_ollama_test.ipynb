```python
import requests
# Replace with the actual IP address of the host machine and port number
url = "http://10.0.6.234:11434/generate"
# Define the prompt you want to send
prompt = {
    "prompt": "Tell me a story about a dragon and a knight."
}
# Send the request to the API
response = requests.post(url, json=prompt)
# Check if the request was successful
if response.status_code == 200:
    # Print the generated text
    print("Generated Text:", response.json().get('response'))
else:
    print(f"Request failed with status code {response.status_code}")
```
